```python
clmns=X.columns
```

```python
clmns
```

```python
def transform(X, fs) :
  X_ = X.copy()
  for i in range(len(fs)) :
    if fs[i] == False :
      X_.drop([clmns[i]], axis=1, inplace=True)
  return X_
```

```python
def Retransform (lst):
  Fs=[]      
  for i in range(225):
    if clmns[i] in lst :
       Fs.append(True)
    else :
       Fs.append(False)
  return Fs
```

```python
lst= ['ZoneAride', 'VegInond', 'VarT_tri_IRM', 'Min_IRM', 'Var_IRM',
       'Moy_TempJ', 'Amp1_TempJ', 'Amp3_TempN', 'VarT_tri_TempN', 'Max_TempN',
       'VarT_ann_NDVI', 'ProfAbs_Sub', 'IR_Moy', 'Prec_Janv', 'Prec_Juill',
       'Rad_Mai', 'Tmax_Avril', 'Vent_Avril', 'Vent_Aout', 'Vent_Sept']
```

```python
Fs = Retransform (lst)
```

```python
Fs
```

```python
cv = KFold(n_splits=5, random_state=1, shuffle=True)
explored_models = []
scores_mcc= []
models_mm=[]
def explor_mcc(fs, mcc, algo) :
  if (sum(fs) < 5 or sum(fs) > 225) :
    return
  explored_models.append(fs)
  for i in range(225) :
    neighbor = fs.copy()
    neighbor[i] = not fs[i]
    if (neighbor not in explored_models) :
      score_mcc = cross_val_score(algo, transform(X, neighbor) , y, scoring=my_scorer1, cv=cv, n_jobs=-1)
      mean_score_mcc = np.mean(score_mcc)
      explored_models.append(neighbor)
      if (mean_score_mcc >= mcc) :
        models_mm.append(neighbor)
        print("mcc", mean_score_mcc)
        print("-----------")
        scores_mcc.append(mean_score_mcc)
        explor_mcc(neighbor, mean_score_mcc, algo)
```
